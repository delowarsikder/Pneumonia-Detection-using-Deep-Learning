```python
from tensorflow.keras.models import load_model
import cv2
import numpy as np
```

```python
from PIL import Image
import glob
image_normal = []
image_pneumonia = []
for filename in glob.glob('../chest_xray//test//NORMAL//*.jpeg'): 
    #im=Image.open(filename)
    im = cv2.imread(filename)
    print(im)
    image_normal.append(im)

for filename in glob.glob('../chest_xray//test//PNEUMONIA//*.jpeg'):
    #im=Image.open(filename)
    im = cv2.imread(filename)
    print(im)
    image_pneumonia.append(im)
```

```python
model = load_model('../chest_xray//xray_train_model.h5')
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
```

```python
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
```

```python
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255)

val_datagen = ImageDataGenerator(rescale=1./255)
```

```python
training_set = train_datagen.flow_from_directory(
        '../chest_xray/train',
        target_size=(64, 64),
        batch_size=32,
        class_mode='binary')
```

```python
test_set = test_datagen.flow_from_directory(
        '../chest_xray/test',
        target_size=(64, 64),
        batch_size=32,
        class_mode='binary')
```

```python
val_set = val_datagen.flow_from_directory(
        '../chest_xray/validation',
        target_size=(64, 64),
        batch_size=32,
        class_mode='binary')
```

```python
# single pic check. normal .
img = cv2.resize(image_normal[33],(64,64))
img = np.reshape(img,[1,64,64,3])
classes = model.predict_classes(img.astype(np.float32))
pred = classes[0][0]
print(type(classes))
print(classes[0][0])
# print(classes.shape)
if(pred==0):
    print ('NORMAL')
else:
    print ('ERROR')
```

```python
# single pic check pnemonia..
img = cv2.resize(image_pneumonia[23],(64,64))
img = np.reshape(img,[1,64,64,3])
classes = model.predict_classes(img.astype(np.float32))
print(type(classes))
print(classes[0][0])
# print(classes.shape)
if(pred==1):
    print ('PNEUMONIA')
else:
    print ('ERROR')
```

```python
#now create two array for real value and the predicted value.....
y_real = []
y_pred = []
```

```python
for image in image_pneumonia : 
    img = cv2.resize(image,(64,64))
    img = np.reshape(img,[1,64,64,3])
    predicted_class = model.predict_classes(img.astype(np.float32))
    y_real.append(1)  
    y_pred.append(predicted_class[0][0])
    

    
for image in image_normal :
    img = cv2.resize(image,(64,64))
    img = np.reshape(img,[1,64,64,3])
    predicted_class = model.predict_classes(img.astype(np.float32))
    y_real.append(0)
    y_pred.append(predicted_class[0][0])
```

```python
print(len(y_real))
print(len(y_pred))
```

```python
print(y_pred)
print(y_real)
```

```python
#chicking the testing accuracy....
from sklearn.metrics import accuracy_score
testing_accuracy = accuracy_score(y_real, y_pred)
print(testing_accuracy)
```
